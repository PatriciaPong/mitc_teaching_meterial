# 迴圈
迴圈就是做重複的事情，以後你想到"重複"請想到迴圈，想到"重複使用"，請想到function。

## While
### 給我一個條件，只要條件符合，我就會跑一輩子。  
這邊不開玩笑，真的是一輩子。你不按Stop，便是永無止境。  

**While格式如下**：
```
while 條件式:
    要重複的事情
```

### 那麼，While要怎麼停下來？  
他條件式的東西不符合了，他就會停下來了。
```
a = 0
while a<10:
    print('a=',end='')
    a = a+1
    print(a)
```
*補充，print(end='')，end後面放的東西，就是這次print最後會補的東西，一般預設是'\\n'也就是換行的意思。*  

你可能會覺得：*啊我看人家都用break，看起來很高科技，我也要我也要！*
- break：常用來跳出迴圈，代表現在重複做的事情立即中斷 
- continue：現在這一圈不做了，下一圈

*來一個很爛的比喻：當你要測驗跑8000，然後你在跑4000之後就去旁邊休息了，這個叫break，如果你跑了4000中途偷偷切西瓜，然後繼續乖乖跑下一圈，這個叫continue*  
break：
```
a = 0
while True:
    print('a=',end='')
    a = a+1
    print(a)
    if a>=10:
        break
```
continue：
```
a = 0
while True:
    a = a+1
    if a%2 == 1:
        continue
    print('a=',end='')
    print(a)
    if a>=10:
        break
```
...說真的，一般人不會寫while True這種沒有彈性的東西。  
除非你有甚麼特殊的想法，不然盡量不要寫這類的奇妙東西。

## 迴圈的習性是，包起來的東西沒跑完，他不會接著往下面跑
```
a = 0
b = 0
while a<10:
    print('a=',end='')
    a = a+1
    print(a)
    while b<10:
        b = b+1
        print('b=',end='')
        print(b)
```
這是最一開始呈現的code，你會發現他中間一大串都是print b，完全沒有print a。  
最後在b已經=10的時候，就不會再進去b的迴圈了，所以只print a。  

## Local／Global
有些人在寫function的時候可能就遇過這個問題了。  
簡單來說：**同樣縮排內的變數，只有在那個縮排內才可以用**（這個基本上只適用python）  
原理是這樣：function內的東西只有function內可以用、迴圈裡面的東西只有迴圈裡面可以用。  

我知道你們聽得很模糊，看個例子：
```
def money():
    give_me_money = '我需要那個酷酷的100元'
    print(give_me_money)
print(give_me_money)
```
當你看到某個變數is not defined，通常是以下原因：
- 變數名稱打錯字
- 前面第一次assign（我們稱做initialize，中文是初始化）沒有run過
- 沒注意到local/global的問題

所有的程式都有一個習慣，如果今天有一個一樣名字的變數已經initialized了，那程式就會當作你要用你之前初始化過的那一個。  
```
give_me_money = '我需要那個酷酷的100元'
def money():
    print(give_me_money)
```

但，當你跑這個東西的時候：
```
give_me_money = 100
def money():
    give_me_money += 100
    print(give_me_money)
money()
```
嘿，有些寫法會告訴你，這個是local variable的give_me_money，但他沒東西可以+=，即便你前面寫了一個give_me_money。  
```
give_me_money = 100
def money():
    global give_me_money
    give_me_money += 100
    print(give_me_money)
money()
```
多補一個global可以提醒他，我們這次要用的就是全域變數

## For
For迴圈比起while常用很多，因為通常有計數類型的迴圈，都會以for的形式表示。  

在教for之前，要先告訴大家**range的用法**：
range(start, end, step)
- 範圍: x會落在start <= x < end的範圍內
- step: 數字之間的間隔，如果是負數，間隔要改成負數

如果沒有end跟step，預設是從0開始數到start；如果沒有step，預設是1。  

**常見For的格式如下：**
```
for 變數 in range(start, end, step):
    要做的事情
```
實務上如果是range，我們習慣把那個變數設成i，如果你今天for迴圈裡面還有一個for迴圈，那麼裡面的變數習慣會設成j，再更內層的通常就是k、m...等等，大家就隨意了。    
實驗看看：
```
for i in range(0,-20,-2):
    print(i)
```
你會注意到，事實上for迴圈不只是「數數」，他有點像是從一袋東西裡面，把裡面的東西一個一個拿出來，等到裡面沒東西了，他就停下來。   
**for實際上的格式如下：**
```
for 變數 in 一袋東西：
    要做的事情
```
那麼接下來，你各位就可以發揮一下自己的想像力。既然只要是一袋東西，list也可以對吧？
```
x = ['a','b','c',1,2,3]
for element in x:
    print(element)
```
如果不是用range，大家就隨意取那個變數的名稱了。　　

### 試試：
想想，各位，想想！還記得我們說過dict要怎麼找出他所有的key嗎？  
這邊有一個現成的dict:
```
money = {
    '1000元':0,
    '500元':0,
    '100元':0,
    '10元':0
}
```
請透過for迴圈，讓使用者連續輸入四次就能設定完他有多少錢，並且每一次輸入的時候都有訊息告訴他現在輸入的是甚麼的數量。

````python
money = {
    '1000元':0,
    '500元':0,
    '100元':0,
    '10元':0
}
for number in money.keys():
    money[number] = input(number)

print(money)
````

### 越來越難了......
除了輸入進去以外，順便累加總金額是多少。

````python
money = {
    '1000':0,
    '500':0,
    '100':0,
    '10':0
}
total = 0
for number in money.keys():
    x = input(number)
    money[number]=x
    total = total +int(number)*int(x)

print(money)
print(total)
````

## 進階迴圈小題目：
大家可能聽過所謂的**演算法**，接下的sort（排序問題）都會有演算法的概念。  
以下題目大家都把最小的放在最左邊（由小至大排列），然後，請不要用python裡面的sort()，不然就沒有意義了(´∩ω∩｀)  

### Selection sort
取好幾次最大值或最小值，一個個放到一個list裡面，阿那個list就是有排序過的了。

### Bubble sort
假如有一個list長這樣：a, b, c, d, e  
a的位置跟b的位置先比，如果b位置的值比a位置的值小，就對調。接著b、c位置比，再來c、d。  
整串都這樣比完之後，再重新回到ab下去比。  
這樣的循環會持續到整串比下來都沒有人要換位置，就會停下來。
