## 除了list之外，還有一個dict
**dict我們稱作dictionary**，就像字典一樣，你有不會的單字，就可以去查他代表什麼。  
而dict就是**一個key（鍵）會對一個value（值）**，key不能重複，但value可以重複。
*想想看，如果我今天要查water是什麼意思，你總不會希望你查到兩個water，然後不知道哪一個是正解，這就是為甚麼key不能重複*
key可以是一個數字、一組中文字，value除了可以是一個數字，他也可以是一組list。

### 查看所有的key：dict.keys()
x.keys()
### 查看所有的value：dict.values()
x.values

## 專有名詞解釋
### Class
即類別，如同字面上意思，幫所有的程式語言中的實體，給他上一個分類，再告訴他可以做甚麼、他有什麼屬性。  
Class底下可以有Class，就像哺乳類底下有人類，人類底下才是我這個人，而我這個人對程式來說，就是一個物件。

### 程式的「動作」：abcd()
- method：方法，寫在Class裡面，就是這類別的物件，可以做的事情。
- function：函式，雖然這個才是函式，但很多人會把method跟function都叫函式或是函數。  
  這個是自己寫的，你可以當作一個程式執行的SOP，你先寫起來放著，以後遇到同樣狀況，就是照著這個SOP去做。

以上兩種都可能會進行一系列的動作，或者傳遞一個結果回來。  
而這兩個東西都可以使用return，回傳一個值然後結束。  
聽不懂？沒關係，我等等會教function怎麼寫。

### 關於物件（Object）的專有名詞
- Object：物件，程式中的個體
- argument：你要塞進去method/function，叫他執行的東西
- parameter：method/function執行的時候需要的東西
- variable：變數，用來存值的東西
- attribute：參數，依據Class不同，會有不同種類的參數，但各個參數是多少，要看每一個物件。

### function從def開始
```
def 名字可以自己取欸(parameter1: type, parameter2: type,...):
```
要叫他什麼，大中天？不不不，我們工作習慣上function都是打英文的。    
type是你預期個別parameter的型別是什麼，用來防呆，比如你今天需要一個整數，你設定int，然後有個傻孩子輸入8.7，jupyter notebook就會幫你報錯。  
parameter1、parameter2...你可以自己取名字，是你接下來會用到的變數名稱。  

那接下來，我們把上面的範例寫成function，以下會列出超多種寫法，但你要記得，我們的目的都是把上面的範例寫成function：
**第一種寫法：function幫我做全套**
```
def money_1():
    x = int(input('1000元的數量：'))
    y = int(input('100元的數量：'))
    z = int(input('50元的數量：'))
    w = int(input('10元的數量：'))
    u = int(input('1元的數量：'))
    print(x*1000+y*100+z*50+w*10+u)
money_1()
```
你會發現：
1. 原來可以不需要parameter
2. 所有放在function裡面，要跑的程式SOP，都要縮排（indent，按下你的tab鍵，不要傻傻按四個空白鍵，很容易手抖）

**第二種寫法：輸入我來，其他交給function**
```
x = int(input('1000元的數量：'))
y = int(input('100元的數量：'))
z = int(input('50元的數量：'))
w = int(input('10元的數量：'))
u = int(input('1元的數量：'))

def money_2(thousand:int, hundred:int, fifty:int, ten:int, one:int):
    print(thousand*1000+hundred*100+fifty*50+ten*10+one)

money_2(x,y,z,w,u)
```
你會發現：
1. 如果今天忘了幫x、y、z、w、u用casting（轉換成int形式），function會出問題。
2. function的parameter名稱，只要function裡面有對到就好，引用function的時候，塞進去的argument不需要跟function的名稱一樣。
3. 不一定要透過input塞x,y,z,w,u，我可以直接使用money_2(10,2,4,5,9)像這樣隨便塞一堆數字，他也會幫我算。
   但要注意，如果塞money_2('10','2',3,4,5)會有紅字，因為你塞進去function的不是你當初設定的type：int。

**第三種寫法：輸出我來，其他交給function**
```
def money_3():
    x = int(input('1000元的數量：'))
    y = int(input('100元的數量：'))
    z = int(input('50元的數量：'))
    w = int(input('10元的數量：'))
    u = int(input('1元的數量：'))
    return x*1000+y*100+z*50+w*10+u
print(money_3())
```
你會發現：
1. return後面不一定要放一個變數，他可以塞整串程式碼，但要記得，return只能回傳「一個東西」，你可以用list包成一個東西，又或者像我一樣，讓他回傳算完之後的那「一個答案」
2. print裡面可以塞function，他會先執行括號裡面的function，再把function回傳的結果print出來

**第四種寫法：我不防呆啦**
```
x = int(input('1000元的數量：'))
y = int(input('100元的數量：'))

    print(thousand*1000+hundred*100+fifty*50+ten*10+one)

```
你會發現：
1. 其實不用跟python講你的parameter type，反正assign變數的時候也不用，function當然也不用
2. 不防呆的情況下，如果你忘記casting，他可能就會做出很恐怖的事情...

## 學會看Document/API，就可以自學囉
以下內容擷取自：https://numpy.org/doc/stable/reference/arrays.ndarray.html  
在各個module的API，經常會看到這種說明：  

| Method | Description |
| --- | --- |
| ```ndarray(shape[, dtype, buffer, offset, …])``` | An array object represents a multidimensional, homogeneous array of fixed-size items.|

這個Method叫做ndarray()，裡面可以放dtype、buffer、offset這些parameter。
ndarray()可以用來表達一個多維度的陣列。*你說我怎麼知道？他Description就這樣寫的。* 

你會注意到，他有說這是一個什麼object，可以推測他會把你塞進去method的東西，轉換成他提到的東西。  
ndarray會把塞進去的東西，轉換成numpy中的array形式，然後回傳（return）。  

如果你仔細去看內文，他會說明dtype、buffer、offset代表什麼意思，像是dtype就是指array裡面放的東西，他的型別是什麼。  

有良心的API就會給你例子：  
```x = np.array([[1, 2, 3], [4, 5, 6]], np.int32)```
你可以看到他把一個長得很像list的東西塞進去array裡面，然後設定一個東西叫做np.int32，你還會注意到np.int32裡面有個很眼熟的東西是int，所以你會推測np.int32應該是numpy自己定義的一種型別（type）。  
接著他把這個method的結果assign給x，因為這個method會回傳一個numpy.ndarray的東西，也就是轉換出來numpy中的array形式。  

這樣看一輪下來，可能你還是不會用這個method，因為你還不夠熟悉。  
但最起碼，你會知道這個method好像會把你的list做甚麼奇妙的轉換，魔法就這麼產生了。
