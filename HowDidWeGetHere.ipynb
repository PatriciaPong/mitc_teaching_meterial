```python
#function
def heal():
    ability['mp'] -= 50
    ability['hp'] += 30
    print('hp->' + str(ability['hp']))
    print('mp->' + str(ability['mp']))
    return
heal()
```

```python
#if-else
##魔法不是無限的，治好了就不用再治了
def heal():
    if ability['hp'] == 500 or ability['mp'] < 50:
        print('魔法發動失效')
        return False
    ability['mp'] -= 50
    ability['hp'] += 30
    print('hp->' + str(ability['hp']))
    print('mp->' + str(ability['mp']))
    return True
heal()
```

```python
#dictionary &  list
##打了幾隻怪，升等了
ability['lv'] = 2
ability['hp'] = 250
ability['mp'] += 200

##獲得新技能
skill.append('快速治療')
```

```python
#for loop
def fast_heal(count:int):
    for i in range(0,count):
        if ability['hp'] == 500 or ability['mp'] < 50:
            print('魔法停止')
            print('{count}次快速治療完成'.format(count=i))
            return
        ability['mp'] -= 50
        ability['hp'] += 30
        print('hp->' + str(ability['hp']))
        print('mp->' + str(ability['mp']))
    print('{count}次快速治療完成'.format(count=count))
```

```python
#input
fast_heal(int(input('請詠唱快速治療的次數：')))
```

```python
#dictionary &  list
##打了幾隻怪，升等了
ability['lv'] = 3
ability['hp'] = 300
ability['mp'] += 200

##獲得新技能
skill.append('完全治療')
```

```python
#while loop
def fully_heal():
    while ability['hp']<500 and ability['mp']>30:
        ability['mp'] -= 50
        ability['hp'] += 30
        print('hp->' + str(ability['hp']))
        print('mp->' + str(ability['mp']))
    print('魔法中止')
    return
fully_heal()
```

```python
#recursive: 自己呼叫自己，又稱遞迴，經典範例為quick sort和河內塔
##這個是補充，以前沒有教過
def fully_heal():
    ability['mp'] -= 50
    ability['hp'] += 30
    print('hp->' + str(ability['hp']))
    print('mp->' + str(ability['mp']))
    if ability['hp']<500 and ability['mp']>30:
        fully_heal()
    else:
        print('魔法中止')
        return  
fully_heal()
```

```python
#import & document
##https://docs.python.org/3/library/json.html
import json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#魔術語言
%matplotlib inline
```

```python
#save & json
def save():
    print('你察覺你的程式魔法正在慢慢進步，這使你充滿了決心......')
    profile = json.dumps({
        'name':name,
        'age':age,
        'skill':skill,
        'package':package,
        'ability':ability
    }, indent=4, ensure_ascii=False)
    #indent是json的排版功能
    #encoding和decoding這件事情很重要，請大家以後遇到亂碼，第一件事情就想到這個
    print('個人資料：',end='')
    print(profile)
    with open('save.json','w',encoding='utf8') as file:
        file.write(profile)
save()
```

### 關於亂碼問題
大家可以試試看把上一個cell的ensure_ascii=False拿掉，你可能會print出奇形怪狀的東西。  
常見的編碼方式有：ascii, big5, utf-8，台灣人的習慣會統一用utf-8。  
所以程式的輸出出現亂碼時：
1. 請檢查你的檔案是什麼編碼
2. 請檢查你程式在處理資料的時候是什麼編碼
3. 請檢查你輸出結果的時候是用什麼編碼

上述三個只要沒有統一，就會有亂碼、奇異的空白出現。

```python
#numpy
X = np.random.rand(100)
noise = np.random.randn(100)
Y = 3*X+60
Y_noise = Y+1.5*noise
```

```python
#matplotlib疑難雜症
plt.rcParams['axes.unicode_minus'] = False #處理負號無法顯示問題
plt.rcParams['font.sans-serif'] = ['simhei']
```

```python
#numpy & matplotlib
plt.scatter(X,Y_noise,color = '#19CAAD',)
plt.xlabel('和森林入口的距離')
plt.ylabel('魔物的數量')
plt.title('神秘森林中的魔物分布傾向')
```
